**Linear regression with one variable**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
path = os.getcwd() + '/ex1data1.txt'
data = pd.read_csv(path, header=None, names=['Population', 'Profit'])
```

**Ex. 2.** Extract data from pandas structure to new variables of ndarray type and plot the data, where: x - population, y - profit.

```python
x = data.values[:, 0]
y = data.values[:, 1]
```

```python
plt.scatter(x, y)
plt.title("Data plot")
plt.xlabel("Population of City in 10.0000s")
plt.ylabel("Profit in $10.0000s")
plt.show()
```

**Ex. 3.** In order to make the cost function work (intercept value θ has to be multiplied by 1) we need to insert a column of ones at the top of x.

**Ex. 4.** We need to separate our data into independent variables X and our dependent variable y. We also need to initialise parameter θ.

```python
X = np.stack([np.ones(x.shape, dtype=np.float32), x], axis=0)
y.shape = [1, y.shape[0]]
theta = np.zeros((X.shape[0], 1), dtype=np.float32)
```

**Ex. 5.** The cost function evaluates the quality of our model by calculating the error between our model’s prediction for a data point, using the model parameters, and the actual data point. Please write the cost function based on 
X, y and parameter θ.

```python
def computeCost(X, y, theta):
  m = y.shape[1]
  return np.float64(np.sum((np.transpose(theta)@X - y)**2)/(2*m))
```

**Ex. 6.** Test the cost function with (θ=0).

```python
computeCost(X, y, theta)
```

**Ex. 7.** Please implement the gradient descent algorithm: 
* the function returns optimized values θ and a vector of the cost function value for each iteration

```python
def simple_gradient(X, y, theta, alpha, it):
    theta = np.float64(theta)
    cost = []
    m = X.shape[1]
    for i in range(it):
        theta = (theta - alpha * (np.transpose((np.transpose(theta)@X-y)@X.T))/m)
        cost.append(computeCost(X, y, theta))
    
    return theta, cost
```

**Ex. 8.** Please calculate the optimal parameters for the dataset.

```python
alpha = 0.01
it = 1000
theta, cost = simple_gradient(X, y, theta, alpha, it)
```

```python
print(theta)
```

**Ex. 9.** Please calculate the cost function for the given θ (Ex.8).

```python
print(cost[-1])
```

**Ex. 10.** Plot the regression function and our dataset.

```python
h1 = np.transpose(theta)@X

plt.scatter(x, y, label="Training data")
plt.plot(x, h1[0], 'r', label="Linear regression")
plt.title("The regression function")
plt.xlabel("Population of City in 10.0000s")
plt.ylabel("Profit in $10.0000s")
plt.legend()
plt.show()
```

**Ex. 11.** Plot the relationship between the cost function and number of iterations.

```python
plt.plot(range(len(cost)), cost)
plt.title("Relationship between the cost function and number of iterations")
plt.xlabel("Iteration number")
plt.ylabel("Cost function value")
plt.show()
```

**Linear regression with multiple variables**

```python
path = os.getcwd() + '/ex1data2.txt'
data2 = pd.read_csv(path, header=None, names=['Size', 'Bedrooms', 'Price'])
data2.head()
```

```python
data2.describe()
```

**Ex. 1.** Notice that the scale of the values for each variable is vastly different. A house will typically have 2-5 bedrooms but may have anywhere from hundreds to thousands of square feet. To fix this, we need to do something called “feature normalization”. One way to do this is by subtracting from each value in a feature the mean of that feature, and then dividing by the standard deviation (2 lines of code).

```python
x1 = data2.values[:,0]
x2 = data2.values[:,1]
y = data2.values[:,2]
```

```python
figure, ax = plt.subplots(1, 2, figsize=(15,8))
ax[0].scatter(x1, y)
ax[0].set_xlabel('Size')
ax[0].set_ylabel('Price')
ax[0].set_title("First variable")
ax[1].scatter(x2, y)
ax[1].set_xlabel('Bedrooms')
ax[1].set_ylabel('Price')
ax[1].set_title("Second variable")
plt.show()
```

```python
# feature normalization
def normalization(x):
    x_norm = []
    for i in range(len(x)):
        x_norm.append((x[i]-np.mean(x))/np.std(x))
    return np.array(x_norm)
```

```python
x1_norm = normalization(x1)
x2_norm = normalization(x2)
```

**Ex.2.** Please perform steps 3-11 from the previous task by using the same functions.

```python
X = np.stack([np.ones(x1.shape, dtype=np.float64), x1_norm,x2_norm], axis=0)
y.shape = [1, y.shape[0]]
theta = np.zeros((X.shape[0], 1), dtype=np.float64)
```

```python
computeCost(X, y, theta)
```

```python
alpha = 0.01
it = 1000
theta, cost = simple_gradient(X, y, theta, alpha, it)
```

```python
print(theta)
```

```python
print(cost[-1])
```

```python
h2 = np.transpose(theta)@X

figure, ax = plt.subplots(1, 2, figsize=(15,8))
ax[0].scatter(x1, y)
ax[0].plot(x1, h2[0], 'r')
ax[0].set_xlabel('Size')
ax[0].set_ylabel('Price')
ax[0].set_title("First variable")
ax[1].scatter(x2, y)
ax[1].plot(x2, h2[0],'r')
ax[1].set_xlabel('Bedrooms')
ax[1].set_ylabel('Price')
ax[1].set_title("Second variable")
plt.show()
```

```python
plt.plot(range(len(cost)), cost)
plt.title("The cost function and number of iterations")
plt.xlabel("Iteration number")
plt.ylabel("Cost function value")
plt.show()
```

**Linear regression - Python packages**

```python
import sklearn.linear_model as sklm
import sklearn.model_selection as skms
 
x = data.values[:,0]
y = data.values[:,1]

X_train,X_test,y_train,y_test = skms.train_test_split(x,y,test_size = 0.3)
regr = sklm.LinearRegression(fit_intercept=True,copy_X = True ).fit(X_train[:,None],y_train)
Y_predicted = regr.predict(X_test [:,None])
 
# Regression coefficients (theta)
print('Coefficients: \n', regr.coef_)
 
#  Residual sum of squares error
error = np.mean((Y_predicted - y_test) ** 2)
print("Residual sum of squares: {}".format(error))
```

```python
figure, ax = plt.subplots(1, 2, figsize=(15,8))
ax[0].scatter(x, y, label='Training data')
ax[0].plot(x, h1[0],'r',label='Linear regression')
ax[0].set_xlabel("Population of City in 10.0000s")
ax[0].set_ylabel("Profit in $10.0000s")
ax[0].set_title("Implementation")
ax[0].legend()
ax[1].scatter(x, y, label='Training data')
ax[1].plot(X_test,Y_predicted,'r',label='Linear regression')
ax[1].set_xlabel("Population of City in 10.0000s")
ax[1].set_ylabel("Profit in $10.0000s")
ax[1].set_title("Scikit learn")
ax[1].legend()
plt.show()
```
