<b>Ex. 4.</b> We need to separate our data into independent variables X and our dependent variable y. We also need to initialise parameter Œ∏. 

```python
X = np.stack([np.ones(x.shape, dtype=np.float32), x], axis=0)
y.shape = [1, y.shape[0]]
theta = np.zeros((X.shape[0], 1), dtype=np.float32)
```

<b>Ex. 5.</b> The cost function evaluates the quality of our model by calculating the error between our model‚Äôs prediction for a data point, using the model parameters, and the actual data point. Please write the cost function based on X,y and parameter Œ∏.

```python
def computeCost(X, y, theta):
    f = theta.transpose() @ X
    m = y.shape[1]
    J = 1 / (2 * m) * np.sum(np.square(f - y)) 
    return J
```

<b>Ex. 6</b> Test the cost function with (Œ∏=0).

```python
computeCost(X, y, theta)
```

<b>Ex. 7.</b> Please implement the gradient descent algorithm

```python
def simple_gradient(X, y, theta, alpha, it):
    theta = np.float64(theta)
    cost = []
    m = X.shape[1]
    for i in range(it):
        theta = (theta - alpha * (np.transpose((np.transpose(theta)@X-y)@X.T))/m)
        cost.append(computeCost(X, y, theta))
        
    return theta, cost
```

```python
alpha = 0.01
it = 1000
theta1, cost = simple_gradient(X, y, theta, alpha, it)
theta1
```

<b>Ex. 1.</b> Notice that the scale of the values for each variable is vastly different. A house will typically have 2-5 bedrooms but may have anywhere from hundreds to thousands of square feet. To fix this, we need to do something called ‚Äúfeature normalization‚Äù. One way to do this is by subtracting from each value in a feature the mean of that feature, and then dividing by the standard deviation

```python
def feature_normalisation_col(col):
    col_norm = []
    for i in range(len(col)):
        col_norm.append((col[i]-np.mean(col))/np.std(col))
    return np.array(col_norm)
```

```python
def feature_normalisation(data):
    for column in data:
        data[column] = (data[column] - data[column].mean()) / data[column].std()
```

## Linear regression - Python packages

One variable

```python
import sklearn.linear_model as sklm
import sklearn.model_selection as skms
path = os.getcwd() + '/ex1data1.txt'
data = pd.read_csv(path, header=None, names=['Population', 'Profit'])
 
x = data.values[:,0]
y = data.values[:,1]

X_train,X_test,y_train,y_test = skms.train_test_split(x,y,test_size = 0.3)
regr = sklm.LinearRegression(fit_intercept=True,copy_X = True ).fit(X_train[:,None],y_train)
Y_predicted = regr.predict(X_test [:,None])
 
# Regression coefficients (theta)
print('Coefficients: \n', regr.coef_)
 
#  Residual sum of squares error
error = np.mean((Y_predicted - y_test) ** 2)
print("Residual sum of squares: {}".format(error))
```
